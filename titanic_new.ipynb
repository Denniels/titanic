## 3. Feature Engineering

Crearemos nuevas caracter√≠sticas y mejoraremos las existentes.

### 2.3 An√°lisis de Variables Categ√≥ricas

Las variables categ√≥ricas contienen informaci√≥n crucial sobre las caracter√≠sticas sociales y log√≠sticas del viaje:

- **Pclass**: Refleja la estratificaci√≥n social a bordo
- **Sex**: Crucial debido a la pol√≠tica "mujeres y ni√±os primero"
- **Embarked**: Puede indicar la clase social y nacionalidad
- **Cabin**: Primera letra indica la cubierta

#### üîç Variables derivadas potenciales:
1. T√≠tulos honor√≠ficos de los nombres
2. Grupos familiares
3. Cubiertas de las cabinas
4. Categor√≠as de tarifas

````python
class TitanicFeatureEngineering:
    def __init__(self):
        self.title_mapping = {}
        self.fare_bins = None
        self.age_bins = None
    
    def extract_title(self, name):
        """Extrae el t√≠tulo del nombre"""
        title = name.split(',')[1].split('.')[0].strip()
        if title in ['Mr', 'Mrs', 'Miss', 'Master']:
            return title
        return 'Other'
    
    def create_family_size(self, df):
        """Crea caracter√≠sticas relacionadas con el tama√±o de la familia"""
        df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
        df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
        return df
    
    def create_age_bins(self, df):
        """Crea bins de edad"""
        df['AgeBin'] = pd.qcut(df['Age'], q=4, labels=['Young', 'Adult', 'MiddleAge', 'Senior'])
        return df
    
    def create_fare_bins(self, df):
        """Crea bins de tarifa"""
        df['FareBin'] = pd.qcut(df['Fare'], q=4, labels=['Low', 'Medium', 'High', 'VeryHigh'])
        return df
    
    def extract_cabin_info(self, df):
        """Extrae informaci√≥n de la cabina"""
        df['CabinDeck'] = df['Cabin'].str[0]
        df['CabinDeck'] = df['CabinDeck'].fillna('U')
        return df
    
    def transform(self, df):
        """Aplica todas las transformaciones de feature engineering"""
        # Copia del dataframe
        df = df.copy()
        
        # Extracci√≥n de t√≠tulo
        df['Title'] = df['Name'].apply(self.extract_title)
        
        # Caracter√≠sticas de familia
        df = self.create_family_size(df)
        
        # Bins de edad y tarifa
        df = self.create_age_bins(df)
        df = self.create_fare_bins(df)
        
        # Informaci√≥n de cabina
        df = self.extract_cabin_info(df)
        
        return df

# Aplicar feature engineering
fe = TitanicFeatureEngineering()
train_data_fe = fe.transform(train_data)
test_data_fe = fe.transform(test_data)

# Mostrar las nuevas caracter√≠sticas
print("Nuevas caracter√≠sticas creadas:")
display(train_data_fe.head())
````

## 4. Preprocesamiento de Datos

Crearemos un pipeline de preprocesamiento robusto.

````python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

class TitanicPreprocessor:
    def __init__(self):
        self.numerical_features = ['Age', 'Fare', 'FamilySize']
        self.categorical_features = ['Sex', 'Embarked', 'Title', 'CabinDeck', 'AgeBin', 'FareBin']
        
    def create_pipeline(self):
        # Pipeline para variables num√©ricas
        numerical_pipeline = Pipeline([
            ('imputer', KNNImputer(n_neighbors=5)),
            ('scaler', StandardScaler())
        ])
        
        # Pipeline para variables categ√≥ricas
        categorical_pipeline = Pipeline([
            ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
            ('onehot', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'))
        ])
        
        # Combinaci√≥n de pipelines
        preprocessor = ColumnTransformer(
            transformers=[
                ('num', numerical_pipeline, self.numerical_features),
                ('cat', categorical_pipeline, self.categorical_features)
            ])
        
        return preprocessor

# Crear y ajustar el preprocesador
preprocessor = TitanicPreprocessor().create_pipeline()

# Separar features y target
X_train = train_data_fe.drop(['Survived', 'PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
y_train = train_data_fe['Survived']
X_test = test_data_fe.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

# Ajustar el preprocesador
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Verificar la forma de los datos procesados
print("Forma de los datos de entrenamiento procesados:", X_train_processed.shape)
print("Forma de los datos de prueba procesados:", X_test_processed.shape)
````

## 5. Modelado y Evaluaci√≥n

Implementaremos m√∫ltiples modelos con validaci√≥n cruzada y optimizaci√≥n de hiperpar√°metros.

````python
%pip install xgboost lightgbm --quiet

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from xgboost import XGBClassifier
import lightgbm as lgb

RANDOM_STATE = 158

class TitanicModeling:
    def __init__(self):
        self.models = {
            'logistic': LogisticRegression(random_state=RANDOM_STATE),
            'random_forest': RandomForestClassifier(random_state=RANDOM_STATE),
            'xgboost': XGBClassifier(random_state=RANDOM_STATE),
            'lightgbm': lgb.LGBMClassifier(random_state=RANDOM_STATE)
        }
        
        self.param_grids = {
            'logistic': {
                'C': [0.001, 0.01, 0.1, 1, 10],
                'penalty': ['l1', 'l2'],
                'solver': ['liblinear']
            },
            'random_forest': {
                'n_estimators': [100, 200],
                'max_depth': [None, 10, 20],
                'min_samples_split': [2, 5],
                'min_samples_leaf': [1, 2]
            },
            'xgboost': {
                'n_estimators': [100, 200],
                'max_depth': [3, 5, 7],
                'learning_rate': [0.01, 0.1]
            },
            'lightgbm': {
                'n_estimators': [100, 200],
                'max_depth': [3, 5, 7],
                'learning_rate': [0.01, 0.1]
            }
        }
    
    def train_and_evaluate(self, X, y, cv=5):
        """Entrena y eval√∫a m√∫ltiples modelos usando validaci√≥n cruzada"""
        results = {}
        
        for name, model in self.models.items():
            # B√∫squeda de hiperpar√°metros
            grid_search = GridSearchCV(
                model,
                self.param_grids[name],
                cv=StratifiedKFold(n_splits=cv, shuffle=True, random_state=RANDOM_STATE),
                scoring='accuracy',
                n_jobs=-1
            )
            
            # Ajustar el modelo
            grid_search.fit(X, y)
            
            # Guardar resultados
            results[name] = {
                'best_score': grid_search.best_score_,
                'best_params': grid_search.best_params_,
                'model': grid_search.best_estimator_
            }
            
            print(f"\nResultados para {name}:")
            print(f"Mejor puntuaci√≥n: {grid_search.best_score_:.4f}")
            print(f"Mejores par√°metros: {grid_search.best_params_}")
        
        return results

# Entrenar y evaluar modelos
modeling = TitanicModeling()
results = modeling.train_and_evaluate(X_train_processed, y_train)
````

## 6. An√°lisis de Resultados y Visualizaciones

````python
def plot_model_comparison(results):
    """Visualiza la comparaci√≥n de modelos"""
    models = list(results.keys())
    scores = [results[model]['best_score'] for model in models]
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x=models, y=scores)
    plt.title('Comparaci√≥n de Modelos - Accuracy en Validaci√≥n Cruzada')
    plt.ylim(0.7, 1.0)
    plt.xticks(rotation=45)
    plt.show()

# Comparar modelos
plot_model_comparison(results)

# Seleccionar el mejor modelo
best_model_name = max(results.keys(), key=lambda k: results[k]['best_score'])
best_model = results[best_model_name]['model']
print(f"\nMejor modelo: {best_model_name}")
print(f"Accuracy en validaci√≥n cruzada: {results[best_model_name]['best_score']:.4f}")
````
